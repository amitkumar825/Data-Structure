```python
#2. Explain the key features of lists in Python.

Answer:-
1. Mutable:- Lists can be modified after they are created. You can change, add, or remove elements.

2. Append:- add element to the end of list.

3. Insert:- insert object before index.

4. Extend:- use to append elements from another iterable.

5. Repeatation operation:- The  repeatation operation, also known as  repeatation or repeat, is a fundamental concept in various fields.

6. Concatenate:- Combining two or more strings of characters to form a new string.

7. Membership:- Checks whether a given value is a member of a sequence such as strings, lists, and tuples.

8. Shallow copy:- Value will change with change in other list.

9. Deep copy:- Deep copy will not change with change in other list.

10. Sorted:- Return a new list containing all items from the iterable in ascending order.

11. Index:- Return first index of value.

12. Count:- Return number of occurrences of value.

13. Remove:- Remove first occurrence of value.

14. Sort:- Sort the list in ascending order and return None.

15. Clear:- Remove all items from list.

16. Pop:- Remove and return item at index (default last).

17. List comprehension:- A list comprehension is a syntactic construct available in some programming languages for creating a list based on existing lists.

18. Capitalize:- To capitalize is to put something in capital letters, particularly the first letters.

19. Split:- Return a list of the substrings in the string, using sep as the separator string. 

20. Conditional list comprehension:- Conditional list comprehension in Python is a concise way to create lists based on conditions.

21. Nested list comprehension:- A way to combine not only one, but multiple for loops, if statements and functions into a single line of code.

22. Stack:- The last element added to the stack will be the first one to be removed.

23. Queue:-FIRST IN FIRST OUT (FIFO)
```

```python
#5. Describe the key features of sets and provide examples of their use.

Answer:-

#Sets:- Unordered collections of unique elements. The order doesn't matter, and duplicate entries are not allowed. Sets are useful for checking membership (if an item exists) or finding the intersection/difference between sets.

Use sets and Example:
    
1. Length:- Return the number of items in a container.
Ex- s = {(2, 3, 4), 1, 2, 3}
    len(s)
    Results- 4
    
    
2. Add:- Add an element to a set.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.add(100)
    s
    Results- {(2, 3, 4), 1, 100, 2, 3}
    
    
3. Pop:- Remove and return an arbitrary set element.Raises KeyError if the set is empty.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.pop()
    Results- 1
    
    
4. Remove:- Remove an element from a set; it must be a member.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.remove(2)
    s
    Results- {(2, 3, 4), 1, 3}
    
    
5. Update:- Update a set with the union of itself and others.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.update("2")
    s
    Results- {(2, 3, 4), 1, 2, '2', 3}
    
    
6. Delete:- The del keyword is used to delete objects. 
Ex- s = {(2, 3, 4), 1, 2, 3}
    del s
    s
    Results- NameError: name 's' is not defined
    
    
7. Clear:- Remove all elements from this set.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.clear()
    s
    Results- set()
    
    
8. Discard:- Remove an element from a set if it is a member.
Ex- s = {(2, 3, 4), 1, 2, 3}
    s.discard(2)
    s
    Results- {(2, 3, 4), 1, 3}
    
    
9. Union:- combines elements from sets excluding duplicates.
Ex- s1 = {"hiking", "reading", "coding"}
    s2 = {"coding", "photography", "tavelling"}
    
    s1 | s2
    Results- {'coding', 'hiking', 'photography', 'reading', 'tavelling'}
    
    
10. Intersection:- only common elements between sets.
Ex- s1 = {"hiking", "reading", "coding"}
    s2 = {"coding", "photography", "tavelling"}
    
    s1 & s2
    Results- {'coding'}
    
    
11. Difference:- return the elements that is present in first set and not in second.
Ex- s1 = {"hiking", "reading", "coding"}
    s2 = {"coding", "photography", "tavelling"}
    
    s1 - s2
    Results- {'hiking', 'reading'}
    
    
12. Symmetric difference:- returns elements that are present in either of sets but not in both.
Ex- s1 = {"hiking", "reading", "coding"}
    s2 = {"coding", "photography", "tavelling"}
    
    s1 ^ s2
    Results- {'hiking', 'photography', 'reading', 'tavelling'}
    
    
13. Frozen set>> immutable version of set., can not be changed added or removed.
Ex- my_frozenset = frozenset({1, 2, 3, 4, 5})
    my_frozenset.add(4, 5, 6)
    Results- AttributeError: 'frozenset' object has no attribute 'add'
    
Ex- my_frozenset = frozenset({1, 2, 3, 4, 5})
    my_frozenset.pop()
    Results- AttributeError: 'frozenset' object has no attribute 'pop'
```
